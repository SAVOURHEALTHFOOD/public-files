```python
import pandas as pd

# Load your CSV file
df = pd.read_csv('NutriInfo.csv')

html_output = [
    '<!DOCTYPE html>',
    '<html>',
    '<head>',
    '  <meta charset="UTF-8">',
    '  <meta name="viewport" content="width=device-width, initial-scale=1">',
    '  <title>Grouped Data Table</title>',
    '  <style>',
    '    /* Overall page styling */',
    '    body {',
    '      background-color: black;',
    '      font-family: Arial, sans-serif;',
    '      margin: 0;',
    '      padding: 10px;',
    '      color: white;',
    '      background-image: url("https://raw.githubusercontent.com/SAVOURHEALTHFOOD/public-files/main/Background.jpg");',
    '      background-size: cover;',
    '      background-position: center;',
    '      background-repeat: no-repeat;',
    '      min-height: 100vh;', # /* Ensures the background stretches to fill the viewport */
    '    }',
    '    /* Details (collapsible sections) styling */',
    '    details {',
    '      margin-bottom: 10px;',
    '    }',
    '    /* Summary (header) styling */',
    '    summary {',
    '      font-weight: bold;',
    '      cursor: pointer;',
    '      background-color: rgba(172, 229, 0);', # /* Slightly transparent green */
    '      border: 1px solid white;',               # /* Fully opaque green border */
    '      color: black;',
    '      padding: 8px;',
    '      border-radius: 5px;',
    '      font-size: 16px;',
    '    }',
    '    /* Add spacing between summary and following content */',
    '    details summary + * {',
    '      margin-top: 10px;',
    '    }',
    '    /* Indent nested details */',
    '    details details {',
    '      margin-left: 20px;',
    '    }',
    '    /* Responsive table container to allow horizontal scrolling */',
    '    .table-container {',
    '      overflow-x: auto;',
    '    }',
    '    /* Table styling */',
    '    table {',
    '      border-collapse: collapse;',
    '      width: 100%;',
    '      margin-bottom: 20px;',
    '      font-size: 16px;',
    '      background-color: transparent;',  #/* Transparent table background */
    '      color: white;',
    '    }',
    '    th, td {',
    '      padding: 4px;',
    '      text-align: left;',
    '      border: none;',
    '      background-color: transparent;',  #/* Transparent cells */
    '    }',
    '    /* Add a border under table headings */',
    '    th {',
    '      border-bottom: 2px solid #ace500;',
    '    }',
    '    /* Mobile adjustments */',
    '    @media (max-width: 600px) {',
    '      table {',
    '        font-size: 14px;',
    '      }',
    '      summary {',
    '        padding: 6px;',
    '      }',
    '    }',
    '  </style>',
    '</head>',
    '<body>'
]

# Define which ItemTypes should be grouped further by Size
item_types_with_size = ['Protein Smoothies', 'Creamy Smoothies', 'Non-Dairy Smoothies']

# Outer grouping by ItemType
for item_type, type_group in df.groupby('ItemType'):
    html_output.append(f'<details>')
    html_output.append(f'  <summary>{item_type} ({len(type_group)} items)</summary>')
    
    if item_type in item_types_with_size:
        # Inner grouping by Size for these item types
        for size, size_group in type_group.groupby('Size'):
            html_output.append(f'  <details>')
            html_output.append(f'    <summary>{size} ({len(size_group)} items)</summary>')
            # Drop both 'ItemType' and 'Size' so they don't repeat inside the table
            group_table = size_group.drop(columns=['ItemType', 'Size'], errors='ignore')
            html_output.append('    <div class="table-container">')
            html_output.append(group_table.to_html(index=False, border=0))
            html_output.append('    </div>')
            html_output.append(f'  </details>')
    else:
        # For other ItemTypes, drop both 'ItemType' and 'Size' columns
        cols_to_drop = []
        if 'ItemType' in type_group.columns:
            cols_to_drop.append('ItemType')
        if 'Size' in type_group.columns:
            cols_to_drop.append('Size')
        cleaned_group = type_group.drop(columns=cols_to_drop, errors='ignore')
        html_output.append('  <div class="table-container">')
        html_output.append(cleaned_group.to_html(index=False, border=0))
        html_output.append('  </div>')
    
    html_output.append(f'</details>')

html_output.append('</body>')
html_output.append('</html>')

with open('grouped_table_2.html', 'w', encoding='utf-8') as f:
    f.write('\n'.join(html_output))
```
